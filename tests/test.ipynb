```python
from __future__ import annotations
import io

from xlsx2html import xlsx2html
from html2text import html2text

with open("./Employee-Sample-Data.xlsx", "rb") as doc:
    xlsx_file = io.BytesIO(doc.read())
    out_file = io.StringIO()

    default_langs = "en"
    xlsx2html(xlsx_file, out_file, locale=default_langs)
    out_file.seek(0)
    result_html = out_file.read()

    markdown = html2text(result_html)
    print(markdown)
```

```python
import io
from striprtf.striprtf import rtf_to_text
import mimetypes

def decode_content(content: bytes) -> str:
    try:
        return content.decode("utf-8")
    except UnicodeDecodeError:
        # Fall back measure for non-UTF-8 encoded text files
        return content.decode("ISO-8859-1")
    
def main():
    file_path = "./simple.rtf"
    mime_type, encoding = mimetypes.guess_type(file_path)
    if mime_type is None:
        mime_type = "text/plain"
    with open(file_path) as data:
    
        print("Content type")
   
        content = data.read()
        byte_data = content.encode('utf-8')
 

        return extract_text(byte_data, mime_type)


def extract_text(content: bytes, content_type: str) -> str:

    decoded_content = decode_content(content)

    if content_type == "text/csv":
        pass

    elif content_type in [
        "text/plain",
        "text/html",
        "text/css",
        "text/javascript",
        "text/markdown",
        "text/vtt",
    ]:
        return decoded_content

    elif content_type == "text/xml":
        return decoded_content
    elif content_type == "text/rtf":
        return rtf_to_text(decoded_content)

    return decoded_content

print(main())
```
